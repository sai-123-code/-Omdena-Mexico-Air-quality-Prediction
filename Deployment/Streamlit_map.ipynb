```python
import folium
from folium.plugins import HeatMap
import pandas as pd
import random
import time
from IPython.display import display, clear_output

# Define initial data for stations
data = {
    'Station': [
        'Ajusco Medio', 'Ajusco', 'Benito Juarez', 'Hospital General de México', 'Merced',
        'Miguel Hidalgo', 'Milpa Alta', 'Pedregal', 'San Agustin', 'Santa Fe',
        'Tlalnepantla', 'UAM Iztapalapa', 'Xalostoc'
    ],
    'Lat': [
        19.272222222222, 19.204499182, 19.371666666667, 19.4135, 19.424722222222,
        19.404, 19.1928, 19.325277777778, 19.533055555556, 19.354375, 
        19.529166666667, 19.36, 19.526111111111
    ],
    'Lon': [
        -99.207777777778, -99.254832314, -99.159166666667, -99.149, -99.119722222222,
        -99.202722222222, -99.0238, -99.204166666667, -99.030555555556, -99.25915,
        -99.204722222222, -99.07, -99.0825
    ],
    'AQI': [random.randint(50, 150) for _ in range(13)]  # Initialize random AQI values
}

df = pd.DataFrame(data)

# Function to randomly update AQI values
def update_aqi():
    for i in range(len(df)):
        df.at[i, 'AQI'] = random.randint(50, 150)  # Random AQI between 50 and 150

# Function to create and display the map
def generate_map():
    # Initialize the map
    m = folium.Map(
        location=[19.4326, -99.1332],  # Central location near Mexico City
        zoom_start=10,
        min_zoom=2,
        max_zoom=20
    )

    # Prepare heatmap data
    heatmap_data = [
        [row['Lat'], row['Lon'], row['AQI']] for _, row in df.iterrows()
    ]

    # Add heatmap layer
    HeatMap(
        heatmap_data,
        min_opacity=0.5,
        max_val=150,
        radius=50,
        blur=35,
        gradient={
            0.0: 'green',
            0.5: 'orange',
            1.0: 'red'
        }
    ).add_to(m)

    # Add custom icons for stations
    for i, row in df.iterrows():
        icon_color = 'red' if row['AQI'] > 100 else 'orange' if row['AQI'] > 75 else 'green'
        folium.Marker(
            location=[row['Lat'], row['Lon']],
            popup=f"{row['Station']}<br>AQI: {row['AQI']}",
            icon=folium.Icon(color=icon_color, icon="info-sign")
        ).add_to(m)

    return m

# Continuously update AQI levels and refresh the map
while True:
    clear_output(wait=True)  # Clear previous output
    update_aqi()  # Update AQI values
    m = generate_map()  # Generate updated map
    display(m)  # Display updated map
    time.sleep(450)  # Wait 45 seconds before updating again
```
